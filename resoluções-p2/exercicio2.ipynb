```python
trad = {'UUU':'Phe','CUU':'Leu','UUA':'Leu','AAG':'Lis','UCU':'Ser','UAU':'Tyr','CAA':'Gln'}
```

```python
def traducao_rna(arquivo_entrada, arquivo_saida):
  fA = open(arquivo_entrada,'r')
  cod = fA.readline()
  fim = traduc(str(cod))
  open(arquivo_saida,'w').write(fim)
```

```python
def traduc(codi):
  cod = list(codi)
  c = []

  c = queb(c,cod)

  fim = 0

  for i in c:
	  a = trad.get(i)
	  if fim == 0:
		  fim = a
	  else:
		  fim += "-" + a
  
  return fim
```

```python
def queb(c,cod):
    for i in range (0,len(cod)-2,3):
	    c += [cod[i] + cod[i+1] + cod[i+2]]
    
    return c
```
