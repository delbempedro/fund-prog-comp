```python
fe = open('message.txt','r')
fs = open('traduction.txt', 'w')

m,n = fe.readline().split()
m,n = int(m),int(n)

trad = {}

for i in range (m):
    t = fe.readline().split(' => ')
    trad[t[0]] = t[1]

for i in range(n):
    ftrad = []
    f = fe.readline().split()
    for j in f:
        if j in trad:
            ftrad.append(trad[j])
        else:
            ftrad.append(i)
    #print(ftrad)
    frase = ''
    for k in ftrad:
        k = k.replace('\n','').replace(' ','')
        if frase == '':
            frase = k
        else:
            frase = frase + ' ' + k
    #print(frase)
        
    open('traduction.txt','a').write(frase)
    open('traduction.txt','a').write('\n')

fs.close
fe.close
```
