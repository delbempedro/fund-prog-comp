```python
nb = int(input())
```

```python
listabombas = []
```

```python
bomba = []
```

```python
for i in range(nb):
  y,x,r = input().split()
  y,x,r = int(y),int(x),int(r)
  bomba = [y,x,r]
  listabombas.append(bomba)
```

```python
for bomb in listabombas:
  dir,esq,cim,bai = 0,0,0,0
  if tab[bomb[0]][bomb[1]] == '1':
    tab[bomb[0]][bomb[1]] = '0'
  for ir in range(1,bomb[2]+1):
    if int(bomb[1]+ir) < c:
      if tab[bomb[0]][bomb[1]+ir] == '1' and dir == 0:
        tab[bomb[0]][bomb[1]+ir] = '0'
        dir = 1
    if int(bomb[1]-ir) > -1:
      if tab[bomb[0]][bomb[1]-ir] == '1' and esq == 0:
        tab[bomb[0]][bomb[1]-ir] = '0'
        esq = 1
    if int(bomb[0]+ir) < l:
      if tab[bomb[0]+ir][bomb[1]] == '1' and bai == 0:
        tab[bomb[0]+ir][bomb[1]] = '0'
        bai = 1
    if int(bomb[0]-ir) > -1:
      if tab[bomb[0]-ir][bomb[1]] == '1' and cim == 0:
        tab[bomb[0]-ir][bomb[1]] = '0'
        cim = 1 
```
