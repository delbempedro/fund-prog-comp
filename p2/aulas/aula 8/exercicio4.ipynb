Um dos motivos para aprender recursão é a facilidade de gerar combinações.

Sua missão agora será a de gerar todas as combinações de N lançamentos de moedas e saber quantas dessas têm uma diferença D entre a quantidade de aparições de caras (K)  e coroas (C).

Por exemplo, no lançamento de duas moedas todas as possibilidades são (C,C) , (C,K) , (K,C) e (K,K). Com uma diferença  0 entre quantidades de caras e coroas temos um total de 2 possibilidades, que são (C,K) e (K,C).

Outro exemplo, se fossem 3 moedas, todas as possibilidades seriam (C,C,C), (C,C,K), (C,K,C), (C,K,K), (K,C,C), (K,C,K), (K,K,C) e (K,K,K). No qual em 6 casos ocorre uma diferença 1 entre a quantidade de caras e coroas.

Entrada

Dois inteiros N e D representando a quantidade de moedas e a diferença procurada.

Saída

Um inteiro indicando a quantidade de possibilidades que satisfazem a diferença procurada.
