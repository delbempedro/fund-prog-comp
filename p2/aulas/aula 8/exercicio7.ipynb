Caso nunca tenha usado nada disso, o balde funciona da seguinte maneira: o usuário clica em um pixel da tela e ele "espalha" a nova cor dele por todos pixels da mesma cor original do pixel que estão conectados com o pixel clicado. Pixels conectados sao os vizinhos do pixel clicado, os vizinhos dos vizinhos e assim por diante. A vizinhança de um pixel e composta pelos 8 pixels vizinhos em sua volta:
