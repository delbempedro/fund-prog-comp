Neymarzinho gosta muito de séries numéricas e resolveu criar uma com seu nome. O primeiro valor da série de Neymarzinho é 3, e cada novo número é gerado somando-se ora 4 ora 1 ao valor anterior, de acordo com o exemplo abaixo:

3  7  8  12  13  17  18  22  23  27 ...

Escreva uma função recursiva chamada SerieNeymarzinho que receba como entrada o índice do elemento na sequência e retorne seu valor.

OBS: Considere que o índice do primeiro elemento dessa sequência é 1.

Entrada

Um valor inteiro, o  índice do elemento na sequência

Saída

Um valor inteiro, o valor dentro da série. Para índices fora do range, ou seja, menores ou iguais a zero, a saída será SEM RESPOSTA.

```python
n = int(input())
```

```python
soma = 3
```

```python
add = 1
```

```python
def SerieNeymarzinho(n):
  global add
  global soma
  if add == 1:
    add = 4
  else:
    add = 1
  if n == 1:
    return soma
  else:
    return add + SerieNeymarzinho(n-1)
```

```python
print(SerieNeymarzinho(n))
```
