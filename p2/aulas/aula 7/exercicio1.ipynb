```python
# calcula det de matriz 3x3
def det3(M):
	t0 = M[1][1]*M[2][2] - M[1][2]*M[2][1]
	t1 = M[1][0]*M[2][2] - M[1][2]*M[2][0]
	t2 = M[1][0]*M[2][1] - M[1][1]*M[2][0]
	return M[0][0]*t0 - M[0][1]*t1 + M[0][2]*t2

# Verifica se a matriz é ou não singular
def isSingular(M):
	return abs(det3(M)) < 1e-4

# Multiplica matrizes A*B
def matrixMult(A,B):
	C = []
	for i in range(3):
		linha = []
		for j in range(3):
			cij = 0
			for k in range(3):
				cij += A[i][k]* B[k][j]
			linha.append(cij)
		C.append(linha)
	return C

# Recebe como entrada uma matriz M e uma potencia p
# Retorna M^p
def matrixPower(M, p):
	Mp = [[1,0,0],[0,1,0],[0,0,1]]
	for i in range(p):
		Mp = matrixMult(M,Mp)
	return Mp

# Recebe uma matriz de entrada e imprime seu conteúdo
def printMatrix(M):
	for i in range(3):
		print('%.2f %.2f %.2f' %  (M[i][0],M[i][1],M[i][2]))
```
