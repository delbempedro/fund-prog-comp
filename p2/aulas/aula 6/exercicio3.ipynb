Faça um programa que leia um número n n n de alunos, a lista dos n n n alunos com suas notas de 10 trabalhos e da prova final.
As informações devem ser armazenadas em um dicionário.

O programa deve ler também um número k k k de consultas, e k k k nomes de alunos, em que k≤n k\leq n k≤n . Para todos os alunos consultados, o programa deve imprimir, na ordem de consulta, o nome do aluno e a sua nota final, de acordo com a fórmula:

NF=∑i=110NTi10∗NP NF = \frac{\sum_{i=1}^{10} NT_i}{10} * NP NF=10∑i=110​NTi​​∗NP

Em que NF NF NF é a nota final do aluno, NTi NT_i NTi​ é a nota do trabalho (entre 0.0 e 10.0) i i i e NP NP NP é a nota da prova (entre 0.0 e 1.0).

```python
n = int(input())

dic = {}

for i in range (n):
  a = input().split()
  dic[a[0]] = a[1:12]
```

```python
k = int(input())

nom = []

for i in range (k):
  nom.append(input())
```

```python
soma = 0
media = 0

for i in nom:
  soma = 0
  for j in range (10):
    soma += float(dic[i][j])
  media = (soma/10)*float(dic[i][-1])
  media = '%.2f'%media
  print(i + ":" + " " + media)
```
