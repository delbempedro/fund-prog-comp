Faça um programa que receba uma frase SS e conte as ocorrências de cada letra nessa frase.

As contagens devem ser salvas num dicionário com as chaves "vogais" e "consoantes", onde o valor de cada uma dessas chaves é outro dicionário com os caracteres pertencentes ao grupo como chaves e as suas contagens como valores.

A contagem deve ser feita ignorando maiúsculas/minúsculas, e as chaves devem ser todas minúsculas. Também devem ser desconsiderados espaços, dígitos e pontuação.

Ao final imprimir o dicionário diretamente, sem formatação.

```python
dic = {'vogais': {}, 'consoantes': {}}
```

```python
frase = input().lower()
```

```python
for i in frase:
  if i.isalpha():
    if (i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u'):
      if i in dic['vogais']:
        dic['vogais'][i] += 1
      else:
        dic['vogais'][i] = 1
    else:
      if i in dic['consoantes']:
        dic['consoantes'][i] += 1
      else:
        dic['consoantes'][i] = 1
```

```python
print(dic)
```
