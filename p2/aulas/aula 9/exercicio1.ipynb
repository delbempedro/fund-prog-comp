```python
def cmatriz(a):
	C =[]
	for i in range(len(a)):
		linha = [float(x) for x in a[i].split()]
		C.append(linha)
	return C
```

```python
def mult_matrizes(A, B, C):
	fA = open(A, "r")
	dimA = [int(x) for x in fA.readline().split()]
	matrizA = fA.readlines()

	fB = open(B, "r")
	dimB = [int(x) for x in fB.readline().split()]
	matrizB = fB.readlines()

	fC = open(C, "w")

	matrizA = cmatriz(matrizA)
	matrizB = cmatriz(matrizB)

	if dimA[1] != dimB[0]:
		fC.write("Incompat√≠veis")
	else:
		m,n,p = dimA[0],dimB[1],dimA[1]
		fC.write(str(m) + " " + str(n) + "\n")

		for i in range(m):
			linha = [0]*n
			for j in range(n):
				for k in range(p):
					linha[j] += matrizA[i][k]*matrizB[k][j]
			fC.write(" ".join(["%.2f"% x for x in linha]))
			fC.write("\n")
					
	fA.close
	fB.close
	fC.close
```
