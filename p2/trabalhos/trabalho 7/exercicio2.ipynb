As quatro estações do ano variam de acordo com as datas:

Primavera: 21 de setembro até 20 de dezembro
Verão: 21 de dezembro até 20 de março
Outono: 21 de março até 20 de junho
Inverno: 21 de junho até 20 de setembro

Escreva uma função

def estacao_do_ano(data):
que recebe como entrada uma data (dia e mês) e retorna o nome da estação correspondente a essa data.

Entrada

Várias linhas com datas em dois formatos. 

No primeiro formato o dia é escrito como um inteiro e o mês por extenso (em minúsculo com possíveis acentuações). Exemplos:

23 de outubro
10 de janeiro
12 de dezembro
Já no segundo formato o dia e o mês são escritos como inteiros e são separados por uma barra (/).    Exemplos:

23/10
10/01
12/12
Sua função deve ser capaz de receber e interpretar as datas em qualquer um desses formatos.

Uma linha com o valor de -1 indica o fim da entrada.

Saída

Várias Linhas com strings correspondentes as estações.

Primavera
Verão
Outono
Inverno

```python
def estacao_do_ano(data):
    a = 0
    estacao = 0
    datas = data.split()
    if len(datas) != 3:
        datas = data.split('/')

    if len(datas) == 3:
        if datas[2] == "janeiro" or datas[2] == "fevereiro":
            estacao = 'Verão'
        if datas[2] == "março":
            if int(datas[0]) <= 20:
                estacao = 'Verão'
            else:
                estacao = 'Outono'
        if datas[2] == "abril" or datas[2] == "maio":
            estacao = 'Outono'
        if datas[2] == "junho":
            if int(datas[0]) <= 20:
                estacao = 'Outono'
            else:
                estacao = 'Inverno'
        if datas[2] == "julho" or datas[2] == "agosto":
            estacao = 'Inverno'
        if datas[2] == "setembro":
            if int(datas[0]) <= 20:
                estacao = 'Inverno'
            else:
                estacao = 'Primavera'
        if datas[2] == "outubro" or datas[2] == "novembro":
            estacao = 'Primavera'
        if datas[2] == "dezembro":
            if int(datas[0]) <= 20:
                estacao = 'Primavera'
            else:
                estacao = 'Verão'
    else:
        if int(datas[1]) == 1 or int(datas[1]) == 2:
            estacao = 'Verão'
        if int(datas[1]) == 3:
            if int(datas[0]) <= 20:
                estacao = 'Verão'
            else:
                estacao = 'Outono'
        if int(datas[1]) == 4 or int(datas[1]) == 5:
            estacao = 'Outono'
        if int(datas[1]) == 6:
            if int(datas[0]) <= 20:
                estacao = 'Outono'
            else:
                estacao = 'Inverno'
        if int(datas[1]) == 7 or int(datas[1]) == 8:
            estacao = 'Inverno'
        if int(datas[1]) == 9:
            if int(datas[0]) <= 20:
                estacao = 'Inverno'
            else:
                estacao = 'Primavera'
        if int(datas[1]) == 10 or int(datas[1]) == 11:
            estacao = 'Primavera'
        if int(datas[1]) == 12:
            if int(datas[0]) <= 20:
                estacao = 'Primavera'
            else:
                estacao = 'Verão'
    return estacao
```
