Implemente a função

def converte(horario):
que converte horários da notação de 24 horas para a notação de 12 horas. 

Entrada

Horários no formato de 24h, um por linha. Se a linha for igual a -1 o programa finaliza.

Saída

Os horários convertidos para a notação de 12h.

Confira abaixo os exemplos.

```python
def converte(horario):
    horario = list(horario.split(":"))
    horario[0] = int(horario[0])
    horario[1] = int(horario[1])
    if horario[0] > 12:
      horario[0] += -12
      horario.append("P.M")
    elif horario[0] == 0:
      horario[0] += 12
      horario.append("A.M")
    elif horario[0] == 12:
      horario.append("P.M")
    else:
      horario.append("A.M")
    a = str('%.2d'%horario[0])
    b = str('%.2d'%horario[1])
    c = str(horario[2])
    horariof = a + ":" + b + " " + c
    return horariof
```

```python
while True:
    horario = input()
    if horario=='-1':
        break
    else:
        print(converte(horario))
```
