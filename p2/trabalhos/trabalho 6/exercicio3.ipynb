Faça um programa que processe uma sequência de tuplas com origens e destinos de uma companhia aérea, representados pelo código do aeroporto. 

O seu programa deve montar um dicionário com os trechos que não possuem volta direta. Nesse dicionário as chaves são as origens, e os valore são listas em ordem alfabética contendo os destinos que não possuem trecho de volta para a origem.

Uma tupla com apenas uma origem, sem destino indica o final da sequência.

Ao final, imprima o dicionário sem formatação .

Exemplo

No exemplo 1 abaixo temos voo direto de GRU para CNF, mas não temos voo de CNF para GRU. O mesmo para o trecho REC -> SSA, que não possui volta SSA -> REC. Todos os demais trechos possuem ida e volta:
GRU -> GIG e GIG -> GRU
GIG -> VCP e VCP -> GIG.

A entrada ('VCP', ) serve apenas para encerrar a lista e não deve ser considerada.
