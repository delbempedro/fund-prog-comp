Escreva um programa que decodifique mensagens encriptadas. Neste problema, as mensagens serão encriptadas através da permutação das letras no alfabeto, cada letra será associada a alguma outra letra do alfabeto (incluindo ela mesma). No exemplo abaixo,  associamos cada letra com a terceira letra a sua frente:



Produzindo o seguinte alfabeto criptografado:

Normal: ABCDEFGHIJKLMNOPQRSTUVWXYZ

Encriptado: DEFGHIJKLMNOPQRSTUVWXYZABC

Assim, se decodificássemos a seguinte mensagem

R SURIHVVRU HVWD ROKDQGR.
seguindo a codificação acima, chegaríamos a mensagem

O PROFESSOR ESTA OLHANDO.
Entrada

A primeira linha da entrada do problema representa o alfabeto encriptado. A primeira letra se refere a letra 'A', a segunda a letra 'B' e assim por diante, como descrito no exemplo. A segunda linha representa uma mensagem codificada.

Todos alfabetos e palavras estarão sempre em letras maiúsculas, assim como devem ser as respostas. Atencao aos espacos em branco, pontos finais e virgulas.

Saída

Imprima a mensagem decodificada.

#tentativa

DEFGHIJKLMNOPQRSTUVWXYZABC
FRSLDU FRGLJR QDR H ERP.

```python
cod = input()
fra = input()
```

```python
alf = {'A':0,'B':1,'C':2,'D':3,'E':4,'F':5,'G':6,'H':7,'I':8,'J':9,'K':10,'L':11,'M':12,'N':13,'O':14,'P':15,'Q':16,'R':17,'S':18,'T':19,'U':20,'V':21,'W':22,'X':23,'Y':24,'Z':25}
num = {0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J',10:'K',11:'L',12:'M',13:'N',14:'O',15:'P',16:'Q',17:'R',18:'S',19:'T',20:'U',21:'V',22:'W',23:'X',24:'Y',25:'Z'}
```

```python
alf['D']
```

```python
n = alf[cod[0]]
n
```

```python
decod = []
```

```python
for i in fra:
  if i in alf:
    con = alf[i] - n
    if con >= 0:
      let = num[con]
      decod += let
    else:
      con2 = 26 + con
      let2 = num[con2]
      decod += let2
  else:
   decod += i
print(decod)
```

```python
fraf = ''
j = 0
for i in decod:
  if j == 0:
    fraf = i
    j += 1
  else:
    fraf += i
```

```python
print(fraf)
```
