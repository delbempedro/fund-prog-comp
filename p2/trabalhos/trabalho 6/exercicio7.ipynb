Faça um programa que conta o número de buracos em uma palavra. Mas o que é um buraco? Imagine, por exemplo, que as letras "A", "D", "O", "P", "R" possuem apenas um buraco. Da mesma forma, a letra "B" possui dois buracos. Já as letras "C", "E", "F", "K" não possuem buracos. Nós devemos considerar que o número de buracos em uma palavra é igual ao número total de buracos nas letras da palavra.

Entrada

A primeira linha contém um inteiro simples N que indica a quantidade de palavras a serem analisadas. Depois, seguem-se N palavras de teste, uma por linha. Cada palavra contém somente caracteres maiúsculos do alfabeto inglês.

Saída

Para cada palavra de teste, a saída consiste de uma linha contendo o número de buracos encontrado na palavra correspondente.

```python
alf = {'A':1,'B':2,'C':0,'D':1,'E':0,'F':0,'G':0,'H':0,'I':0,'J':0,'K':0,'L':0,'M':0,'N':0,'O':1,'P':1,'Q':1,'R':1,'S':0,'T':0,'U':0,'V':0,'W':0,'X':0,'Y':0,'Z':0}
```

```python
n = int(input())
```

```python
for i in range (n):
  cont = 0
  p = input()
  for j in range (len(p)):
    cont += alf[p[j]]
  print(cont)
```
