Escreva um programa que conte quantas palavras diferentes, quantas palavras repetidas e quantas frases existem em um texto. Uma frase é delimitada por um ponto final, ponto de exclamação ou ponto de interrogação.

Para palavras repetidas, considerar o número de repetições. Uma palavra que aparece:

1 vez é considerada repetida 0 vezes;
2 vezes é considerada repetida 1 vez;
3 vezes é considerada repetida 2 vezes;
4 vezes é considerada repetida 3 vezes;
E assim sucessivamente.

experience whine oranges control harsh suspect, obsequious milk finger savory worthless, salt. spy curve! obsequious texture, ready outrageous office harsh control, haircut admire. texture? experience well, control! tense request finger.

```python
a = input().split('.')
```

```python
b = []
for i in a:
  if not b:
    b = i.split('!')
  else:
    b += i.split('!')
```

```python
c = []

for i in b:
  if not c:
    c = i.split('?')
  else:
    c += i.split('?')
```

```python
t = len(c) - 1
t
```

```python
for i in range (len(c)):
    c[i] = c[i].split(',')
```

```python
for i in range (len(c)):
  for j in range (len(c[i])):
     c[i][j] = c[i][j].split()
```

```python
c
```

```python
control = {}
for i in c:
  for j in i:
    for k in j:
      if k.lower() in control:
        k = k.lower()
        control[k] += 1
      else:
        k = k.lower()
        control[k] = 0
```

```python
control
```

```python
p = 0
r = 0
cc = 0
for i in control:
  if control[i] == 0:
    p += 1
  else:
    r += control[i]
  if control[i] >= 2:
    cc += control[i] - 1
```

```python
x = p+r-cc
print(x, r, t)
```
