Escreva um programa que leia os dados de um dicionário e faça a tradução de frases em inglês para português de acordo com o dicionário fornecido

Entrada

Uma linha contendo um número inteiro (N) de palavras do dicionário. Em seguida, as N linhas que representam o conteúdo do dicionário. Em cada uma destas linhas, há uma palavra seguida da string ‘ => ’ e da palavra para a qual ela deve ser traduzida.

Após carregar os dados do dicionário, o programa deve ler uma sequência de linhas de qualquer tamanho. Estas linhas devem ser traduzidas conforme os dados do dicionário. O programa acaba quando for encontrada uma linha que contém apenas o caractere ‘*’ (asterisco).

Saída

Para cada frase em inglês fornecida na entrada, deve ser impressa uma linha com a tradução correspondente em português.

```python
n = int(input())
```

```python
trad = {}
```

```python
for i in range (n):
  t = input().split(' => ')
  trad[t[0]] = t[1]
```

```python
trad
```

```python
c = False
while c == False:
  ftrad = []
  f = input().split()
  if f == ['*']:
    c = True
  else:
    for i in f:
      if i in trad:
        ftrad.append(trad[i])
      else:
        ftrad.append(i)
    print(*ftrad)
```
