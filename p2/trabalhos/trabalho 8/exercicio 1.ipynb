Em que VV representa o valor de um ingresso básico em reais (R$) e xx representa a quantidade de pessoas que pagaram ingresso.

Sua tarefa é fazer um programa que dado o valor do ingresso básico calcule a arrecadação total da partida.

Entrada

Quantidade de pessoas que pagaram ingresso e valor do ingresso básico.

Saída

Valor total da arrecadação.

```python
def arrecad(x,v):
  if x == 0:
    return -100000
  else:
    if int(x/2) == x/2:
      return x + arrecad(x-100,v)
    else:
      return v + arrecad(x-100,v)
```

```python
n, v = input().split()
n, v = int(n), float(v)
```

```python
v_final = arrecad(n,v)
currency = "{:,}".format(v_final)
cur = currency.replace(",",".")
print("R$ "+cur+",00")
```
