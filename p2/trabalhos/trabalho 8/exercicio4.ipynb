```python
qtd_moedas,d=input().split()
qtd_moedas,d=int(qtd_moedas),int(d)
qtd_casos=2**qtd_moedas
def fibo(quantidade_moedas):
    m=[]
    for i in range(quantidade_moedas):
        v=[0]*(quantidade_moedas+1)
        m.append(v)
    m[0][0]=1
    m[0][1]=1

    for i in range(len(m)):
        for j in range(len(m[i])):
            if j==0:
                m[i][j]=1
            elif i!=0 and j!=0:
                m[i][j]=m[i-1][j]+m[i-1][j-1]
    return m[len(m)-1]
fibs=fibo(qtd_moedas)
if qtd_moedas%2==1:
    for i in range(0,qtd_moedas):
        a=qtd_moedas-2*i
        if d==a:
            print(2*fibs[i])
            break
        elif d%2==0:
            print(0)
            break
else:

    for i in range(0,qtd_moedas):
        a=qtd_moedas-2*i
        if d==a:
            print(2*fibs[i])
            break
        elif d==0:
            print(fibs[len(fibs)//2])
            break
        elif d%2==1:
            print(0)
            break
```
