```python
def print_dict_rec(D,level=0):
    tabs = '\t'*level
    for chave in list(D.keys()):
        if not isinstance(D[chave],dict):
            print('%s%s: %s' % (tabs, chave, str(D[chave])))
        else:
            print('%s%s: {' % (tabs, chave))
            print_dict_rec(D[chave],level+1)
            print('%s}' % tabs)

D = eval(input().strip())
print_dict_rec(D)
```
