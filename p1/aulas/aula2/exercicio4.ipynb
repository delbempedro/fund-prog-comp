```python
massa, altura = input().split()
massa, altura = float(massa), float(altura)
```

```python
imc = massa/(altura**2)
```

```python
print("%.2f"%imc)
if imc < 18.5:
  print("abaixo do peso")
elif 18.5 <= imc < 24.9:
  print("saudável")
elif 24.9 <= imc < 29.9:
  print("sobrepeso")
elif 29.9 <= imc < 34.9:
  print("obesidade grau 1")
elif 34.9 <= imc < 39.9:
  print("obesidade severa")
elif 40.0 <= imc:
  print("obesidade mórbida")
```
