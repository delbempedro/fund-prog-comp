```python
x1,y1,r1 = input().split()
x1,y1,r1 = float(x1), float(y1), float(r1)
```

```python
x2,y2,r2 = input().split()
x2,y2,r2 = float(x2), float(y2), float(r2)
```

```python
X = ((x1 - x2)**2)**(0.5)
Y = ((y1 - y2)**2)**(0.5)
soma = X**2 + Y**2
distancia = soma**(0.5)
raios = r1 + r2
```

```python
if distancia <= raios:
  print ("S")
else:
  print("N")
```
