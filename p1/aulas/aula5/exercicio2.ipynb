Faça um programa que leia a ordem m \times nm×n  de uma matriz AA e calcule a norma de Fröbenius dessa matriz, dada por:

```python
linhas,colunas = input().split()
linhas,colunas = int(linhas),int(colunas)
```

```python
matriz = []
```

```python
for i in range (linhas):
	linha = [float(i) for i in input().split()]
	matriz.append(linha)
```

```python
result = 0
```

```python
for i in range (linhas):
	for j in range (colunas):
		result += (matriz[i][j])**2
```

```python
result = result**(1/2)
```

```python
print("%.2f"%result)
```
