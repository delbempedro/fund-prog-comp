  e calcule o produto escalar entre eles.

```python
vetor1 = [float(i) for i in input().split()]
vetor2 = [float(j) for j in input().split()]
```

```python
result = 0
```

```python
for k in range(len(vetor1)):
	result = result + vetor1[k]*vetor2[k]
```

```python
print("%.2f"%result)
```
