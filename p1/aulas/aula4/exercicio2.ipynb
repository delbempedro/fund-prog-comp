Faça um programa que leia um vetor de números inteiros com 10 posições, e conte quantos valores diferentes existem no vetor.

#tentativa em casa

```python
vetor = [int(i) for i in input().split()]

dif = 10
vetorc = []
a = 0

for k in range(0, 10):
	for j in range(k, 10):
		if k != j:
			if vetor[k] == vetor[j]:
				if a == 0:
					dif = dif - 1
					vetorc.append(vetor[k])
					a = 1
				else:
					for l in range (0, len(vetorc)):
						if vetorc[l] !=  vetor[k]:
							dif = dif - 1
							vetorc.append(vetor[k])

print(dif)
```
