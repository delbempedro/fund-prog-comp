#tentativa em aula

```python
vetor1 = [i for i in input().split()]
vetor2 = [j for j in input().split()]

vetorf = []

menor = 0
maior = 0
t = 0

if len(vetor1) < len(vetor2):
	menor  = 1
elif len(vetor2) < len(vetor1):
	maior  = 1

	
if menor == 1:
	t = len(vetor2) - len(vetor1)
if maior == 1:
	t = len(vetor1) - len(vetor2)

if menor == 1:
	for k in range (len(vetor1)):
		vetorf.append(vetor1[k])
		vetorf.append(vetor2[k])
if maior == 1:
	for k in range (len(vetor2)):
		vetorf.append(vetor1[k])
		vetorf.append(vetor2[k])
	
if menor == 1:
	for l in range (len(vetor1),len(vetor2)):
		vetorf.append(vetor2[l])
if maior == 1:
	for l in range (len(vetor2),len(vetor1)):
		vetorf.append(vetor1[l])

f = ' '.join(vetorf)

print(f)
```

```python
l1 = input().split()
l2 = input().split()

r = []

n1,n2 = len(l1),len(l2)
n = min(n1,n2)

for i in range(n):
	e1 = l1.pop(0)
	e2 = l1.pop(0)
	r.append(e1)
	r.append(e2)
	
if len(l1):
	r += l1
else:
	r += l2
resp = " ".join(r)    
print(resp)
```
