Faça um programa que lista todos os divisores de um número natural n \geq 2n≥2 ou diz que o número é primo caso não existam divisores além de 11 e nn.

```python
n = int(input())
```

```python
primo = 1
```

```python
for i in range (2, n):
	if n%i == 0:
		print(i)
		primo = 0
```

```python
if primo == 1:
	print("primo")
```
