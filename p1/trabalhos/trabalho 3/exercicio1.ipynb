Faça um programa que imprima a tabuada de multiplicação de um número inteiro positivo nn fornecido.

```python
n = int(input())
```

```python
for i in range(1,11):
	produto = n*i
	n = str(n)
	i = str(i)
	produto = str(produto)
	print(n + " " + "*" + " " + i + " " + "=" + " " + produto)
	n = int(n)
	i = int(i)
	produto = int(produto)
```
