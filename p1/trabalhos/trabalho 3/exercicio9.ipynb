Escreva um programa que leia um inteiro nn seguido por uma sequência de n + 1n+1 linhas. Destas n + 1n+1 linhas, a última é uma string contendo caracteres cujas ocorrências nas nn linhas anteriores devem ser contadas e impressas individualmente na saída do programa.

Entrada

Um número inteiro positivo nn, seguido por  n + 1n+1 strings. 

Saída

kk linhas, onde kk corresponde ao número de caracteres na última string da entrada. Cada linha deve conter o caractere seguido por um espaço em branco, o sinal de igual (=), outro espaço em branco e um número inteiro que corresponda ao número de vezes que o caractere foi encontrado nas nn linhas anteriores.

```python
n = int(input())
```

```python
frase = input()
find = None
```

```python
for i in range (1, n+1):
	if i < n:
		frase = frase + input()
	else:
		find = input()
```

```python
a = len(find)
for j in range (0, a):
	b = find[j]
	c = frase.count(b)
	
	b = str(b)
	c = str(c)
	print(b +  " " + "=" + " " + c)
```
