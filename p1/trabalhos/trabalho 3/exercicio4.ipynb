Faça um programa que imprima os horários de um ônibus, sabendo que ele passa a cada tt minutos e que as linhas começam a funcionar às 5:00 (primeiro horário) e param às 23:55 (último horário possível).

Para imprimir as horas e minutos sempre com 2 dígitos, incluindo zeros à esquerda, utilize o comando:

print('%02d:%02d' % (horas, minutos))

```python
t = int(input())
```

```python
horas, minutos = 5, 0
```

```python
while horas < 24:
	print('%02d:%02d' % (horas, minutos))
	if 60 <= minutos + t:
		minutos = t -(60 - minutos)
		horas += 1
	else:
		minutos = minutos + t
	if t > 60:
		while 60 <= minutos:
			horas += 1
			minutos = minutos - 60
```
