Neste problema em específico, use 48 casas decimais.

Note que o somatório é definido para k=0, \dots, \inftyk=0,…,∞. Para uma aproximação com um número fixo de termos ele passa a ser k=0, \dots, nk=0,…,n.
Como os termos são valores muito próximos da precisão de armazenamento de números reais em um computador, a ordem da soma pode interferir no resultado devido à propagação do erro de representação ao somar cada termo.
Para evitar erros desse tipo, siga a ordem de k=0, \dots, nk=0,…,n. Evite somar os termos na ordem inversa (k=n, \dots, 0k=n,…,0).


```python
n = int(input())
```

```python
resultado = 0
```

```python
for k in range (0, n+1):
	formula = (1/16**k)*((4/(8*k + 1)) - (2/(8*k + 4)) - (1/(8*k + 5)) - (1/(8*k + 6)))
	resultado = resultado + formula
```

```python
print("%.48f"%resultado)
```
