Escreva um programa que

peça uma nota entre 0 e 10; 
verifica se a nota inserida está entre 0 e 10;
mostra a mensagem "nota válida", caso o valor inserido seja válido, ou continua pedindo uma nota, caso a nota inserida não seja válida.

```python
print("Informe uma nota entre 0 e 10!")
```

```python
nota = int(input())
```

```python
while nota < 0 or nota > 10:
	print("Nota inválida, digite uma nota entre 0 e 10!")
	nota = int(input())
```

```python
print("Nota válida!!!")
```
