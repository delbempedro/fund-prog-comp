```python
n = int(input())
```

```python
c9 = int(n/10**8)
c8 = int(n/10**7) - (10**1)*c9
c7 = int(n/10**6) - (10**2)*c9 - (10**1)*c8
c6 = int(n/10**5) - (10**3)*c9 - (10**2)*c8 - (10**1)*c7
c5 = int(n/10**4) - (10**4)*c9 - (10**3)*c8 - (10**2)*c7 - (10**1)*c6
c4 = int(n/10**3) - (10**5)*c9 - (10**4)*c8 - (10**3)*c7 - (10**2)*c6 - (10**1)*c5
c3 = int(n/10**2) - (10**6)*c9 - (10**5)*c8 - (10**4)*c7 - (10**3)*c6 - (10**2)*c5 - (10**1)*c4
c2 = int(n/10**1) - (10**7)*c9 - (10**6)*c8 - (10**5)*c7 - (10**4)*c6 - (10**3)*c5 - (10**2)*c4 - (10**1)*c3
c1 = int(n/10**0) - (10**8)*c9 - (10**7)*c8 - (10**6)*c7 - (10**5)*c6 - (10**4)*c5 - (10**3)*c4 - (10**2)*c3 - (10**1)*c2
```

```python
if c9 != 0:
	e = 9
elif c8 != 0:
	e = 8
elif c7 != 0:
	e = 7
elif c6 != 0:
	e = 6
elif c5 != 0:
	e = 5
elif c4 != 0:
	e = 4
elif c3 != 0:
	e = 3
elif c2 != 0:
	e = 2
elif c1 != 0:
	e = 1
```

```python
resultado = c9**e + c8**e + c7**e + c6**e + c5**e + c4**e + c3**e + c2**e + c1**e
```

```python
if n == resultado:
	print("True")
else:
	print("False")
```
