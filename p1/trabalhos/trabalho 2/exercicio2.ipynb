Faça um programa que receba um número inteiro e exiba uma mensagem indicando se este número é par ou ímpar, se é positivo ou negativo, ou se é zero.

```python
n = input()
n = float(n)
```

```python
x = n/2
a = 2*x
b = 2*int(x)
if a == b:
  print("Par")
if a != b:
  print("Impar")
```

```python
if n < 0:
  print("Negativo")
if n > 0:
  print("Positivo")
if n == 0:
  print("Zero")
```
