Faça um programa que leia três coordenadas em R2
  e indique se formam um triângulo, juntamente com o seu tipo: equilátero, isósceles ou escaleno.

```python
x1, y1 = input().split()
x2, y2 = input().split()
x3, y3 = input().split()
x1,x2,x3,y1,y2,y3 = float(x1),float(x2),float(x3),float(y1),float(y2),float(y3)
```

```python
X1 = ((x1 - x2)**2)**(0.5)
X2 = ((x1 - x3)**2)**(0.5)
X3 = ((x2 - x3)**2)**(0.5)
Y1 = ((y1 - y2)**2)**(0.5)
Y2 = ((y1 - y3)**2)**(0.5)
Y3 = ((y2 - y3)**2)**(0.5)
```

```python
l1 = (X1**2 + Y1**2)**(0.5)
l2 = (X2**2 + Y2**2)**(0.5)
l3 = (X3**2 + Y3**2)**(0.5)
```

```python
cond1 = l1 + l2
cond2 = l1 + l3
cond3 = l2 + l3
```

```python
cond4 = 0
naotriangulo = 0
```

```python
if cond1 > l3 and cond2 > l2 and cond3 > l1:
  print("triângulo")
  cond4 = 1
if cond4 == 0:
  print("não triângulo")
  naotriangulo = 1
```

```python
cond5 = 0
```

```python
if naotriangulo == 0:
  if l1 == l2 == l3:
    print("equilátero")
    cond5 = 1
  if l1 == l2 or l1 == l3 or l2 == l3 and cond5 == 0:
    print("isósceles")
    cond5 = 2
  if cond5 == 0:
    print("escaleno")
```
