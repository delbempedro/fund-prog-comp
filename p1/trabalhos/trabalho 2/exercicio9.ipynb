Vermelho, azul e amarelo são chamadas de cores primárias porque não podem ser obtidas pela mistura de outras cores. Quando duas cores primárias são misturadas, uma cor secundária é obtida. Por exemplo, se você misturar:

vermelho e azul, obterá roxo;
vermelho e amarelo, obterá laranja;
azul e amarelo, obterá verde.
Escreva um programa que leia os nomes de duas cores primárias que serão misturadas. Se o usuário inserir qualquer coisa diferente de vermelho, azul ou amarelo, o programa deve exibir uma mensagem de erro. Caso contrário, o programa deve imprimir o nome da cor resultante.

Observação: Se você misturar vermelho e vermelho, obterá vermelho. Vale o análogo para as demais cores primárias.

```python
cor1 = input()
cor2 = input()
```

```python
printado = 0
```

```python
if cor1 == cor2:
  if cor1 == "vermelho" and printado == 0:
    print("vermelho")
    printado = 1
  if cor1 == "azul" and printado == 0:
    print("azul")
    printado = 1
  if cor1 == "amarelo" and printado == 0:
    print("amarelo")
    printado = 1
else:
  if (cor1 == "vermelho" and cor2 == "azul" and printado == 0) or (cor2 == "vermelho" and cor1 == "azul" and printado == 0):
    print("roxo")
    printado = 1
  if (cor1 == "vermelho" and cor2 == "amarelo" and printado == 0) or (cor2 == "vermelho" and cor1 == "amarelo" and printado == 0):
    print("laranja")
    printado = 1
  if (cor1 == "amarelo" and cor2 == "azul" and printado == 0) or (cor2 == "amarelo" and cor1 == "azul" and printado == 0):
    print("verde")
    printado = 1
if printado == 0:
  print("cores não primárias")
```
