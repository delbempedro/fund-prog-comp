Escreva um programa que determine se um determinado ano é um ano bissexto. Um ano é um ano bissexto se seu número for um múltiplo de 4, mas não um múltiplo de 100, ou se for um múltiplo de 400.

```python
ano = int(input())
```

```python
a = ano/4
a = int(a)
a = a*4
divisivelpor4 = 0
if a == ano:
  divisivelpor4 = 1
```

```python
a = ano/100
a = int(a)
a = a*100
divisivelpor100 = 0
if a == ano:
  divisivelpor100 = 1
```

```python
a = ano/400
a = int(a)
a = a*400
divisivelpor400 = 0
if a == ano:
  divisivelpor400 = 1
```

```python
bissexto = 0
```

```python
if (divisivelpor4 == 1 and divisivelpor100 == 0) or (divisivelpor400 == 1):
  print("SIM")
  bissexto = 1
```

```python
if bissexto == 0:
  print("NÃO")
```
