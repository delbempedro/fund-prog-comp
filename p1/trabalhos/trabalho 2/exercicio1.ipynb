Faça um programa que calcule as raízes reais de uma equação do segundo grau
f(x) = a*x^2 + b*x + c usando a fórmula de Bhaskara.

Bhaskara:
x = (-b ± √Δ)/2

Δ = b^2 - 4*a*c

```python
a, b, c = input().split()
a, b, c = float(a), float(b), float(c)
```

```python
delta = b**2 - 4*a*c
```

```python
realounao = 0
```

```python
if delta < 0:
  realounao = 1
  print ("No real roots.")
```

```python
if realounao == 0:
  raiz1 = (-b + (delta)**(0.5))/(2*a)
  raiz2 = (-b - (delta)**(0.5))/(2*a)
  if raiz1 == raiz2:
   print("%.2f"%raiz1)
  if raiz1 != raiz2:
   print("%.2f"%raiz1)
   print("%.2f"%raiz2)
```
