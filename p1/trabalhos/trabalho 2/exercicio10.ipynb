Na roleta, os bolsos são numerados de 0 a 36. Abaixo estão as cores dos bolsos:

bolso 0 verde; para os bolsos de 1 a 10, os bolsos ímpares são vermelhos e os bolsos pares são pretos; para os bolsos 11 a 18, os bolsos ímpares são pretos e os bolsos pares são vermelhos; para os bolsos de 19 a 28, os bolsos ímpares são vermelhos e os bolsos pares são pretos; para os bolsos de 29 a 36, ​​os bolsos ímpares são pretos e os pares, vermelhos. Escreva um programa que leia o número do bolso e indique se o bolso é verde, vermelho ou preto. O programa deve exibir uma mensagem de erro se o usuário inserir um número fora do intervalo de 0 a 36.

```python
n = input()
n = float(n)
```

```python
x = n/2
a = 2*x
b = 2*int(x)
if a == b:
  decisor = "par"
if a != b:
  decisor = "impar"
```

```python
if n < 0 or n > 36:
  print("entrada inválida")
if 1 <= n <= 10:
  if decisor == "impar":
    print("vermelho")
  else:
    print("preto")
if 11 <= n <= 18:
  if decisor == "par":
    print("vermelho")
  else:
    print("preto")
if 19 <= n <= 28:
  if decisor == "impar":
    print("vermelho")
  else:
    print("preto")
if 29 <= n <= 36:
  if decisor == "par":
    print("vermelho")
  else:
    print("preto")
if n == 0:
  print("verde")
```
